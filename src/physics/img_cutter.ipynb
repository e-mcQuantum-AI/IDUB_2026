```python
import cv2
import math
import matplotlib.pyplot as plt
from pathlib import Path

base_path = Path.cwd() 
two_back = base_path.parents[1]
img_path = two_back / "quantum_dataset" / "clean" / "coherent_a1.62_id1.png"
img_bgr = cv2.imread(str(img_path))

img = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

rozmiar = 10
kawalki = []

for r in range(0, img.shape[0], rozmiar):
    for c in range(0, img.shape[1], rozmiar):
        fragment = img[r:r+rozmiar, c:c+rozmiar]
        kawalki.append(fragment)

ile_kawalkow = len(kawalki)
kolumny = rozmiar
wiersze = (ile_kawalkow // kolumny) + 1

plt.figure(figsize=(10, 10))
for i, kawalek in enumerate(kawalki):
    plt.subplot(wiersze, kolumny, i + 1)
    plt.imshow(kawalek)
    plt.axis('off')

plt.show()
```
